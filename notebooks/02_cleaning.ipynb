# 02 — Data Cleaning

Show the raw CDC quirk (two spellings of 'Black') and save the tidy tables.

**Outputs**: `data/processed/`

```python
# add the project folder to the path so we can import our code in src/
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
```

```python
from src.data.sources import cdc
from src.config import PROCESSED_DIR
```

```python
# raw data: the same group is spelled 'Black' and 'Black ' (trailing space)
raw = cdc.load()
raw['race'].value_counts()
```

```python
# cdc.tidy() strips the whitespace and standardizes the schema
df = cdc.tidy(raw)
df['entity'].value_counts()
```

```python
# one row per year, one column per race
wide = cdc.to_wide(df)
wide.head()
```

```python
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
df.to_csv(PROCESSED_DIR / 'cdc_clean.csv', index=False)
wide.to_csv(PROCESSED_DIR / 'cdc_wide.csv')
list(PROCESSED_DIR.glob('*.csv'))
```
