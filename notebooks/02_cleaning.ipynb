Fix the race-label whitespace, set column types, and save the tidy long and wide tables.

**Input**: raw CDC pull · **Outputs**: `data/processed/`

```python
# add the project folder to the path so we can import our code in src/
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
```

```python
import pandas as pd

from src.data.load import load_raw
from src.data.clean import clean, to_wide
from src.config import PROCESSED_DIR
```

```python
# the raw data spells the same group two ways: 'Black' and 'Black ' (trailing space)
df = load_raw()
df['race'].value_counts()
```

```python
# strip the whitespace and fix types
df = clean(df)
df['race'].value_counts()
```

```python
# pivot to one row per year, one column per race
wide = to_wide(df)
wide.head()
```

```python
# save both shapes for the next notebooks
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
df.to_csv(PROCESSED_DIR / 'cleaned.csv', index=False)
wide.to_csv(PROCESSED_DIR / 'cleaned_wide.csv')
list(PROCESSED_DIR.glob('*.csv'))
```
