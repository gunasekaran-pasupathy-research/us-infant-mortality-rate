# 03 — Statistical Testing (the Black–White disparity)

Quantify and test the gap, and measure how strongly the rate tracks the year.

**Input**: CDC tidy data

```python
# add the project folder to the path so we can import our code in src/
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
```

```python
import numpy as np
from scipy import stats

from src.data.sources import cdc
```

```python
# Black-minus-White gap and ratio for every year
df = cdc.tidy()
gap = cdc.black_white_gap(df)
gap.describe().round(2)
```

```python
# each year pairs a Black rate with a White rate -> paired t-test on the disparity
t, p = stats.ttest_rel(gap['black'], gap['white'])
print(f'paired t = {t:.2f}, p = {p:.3e}')
print(f'mean gap = {gap["difference"].mean():.2f} per 1,000')
print(f'mean ratio = {gap["ratio"].mean():.2f}x')
```

```python
# how strongly does the rate move with the year within each race?
for race, g in df.groupby('entity'):
    r, p = stats.pearsonr(g['year'], g['imr'])
    print(f'{race:6} Pearson r = {r:.3f}, p = {p:.2e}')
```

## Notes

- The gap is large, persistent, and highly significant.
- Year vs rate is strongly negative but curved — a linear r understates it (notebook 04).
