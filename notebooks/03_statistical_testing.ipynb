# 03 — Statistical Testing

Quantify and test the Black–White disparity, and measure how strongly the rate tracks the year.

**Input**: `data/processed/cleaned.csv`

```python
# add the project folder to the path so we can import our code in src/
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
```

```python
import pandas as pd
import numpy as np
from scipy import stats

from src.config import PROCESSED_DIR, RATE_COL
from src.data.clean import black_white_gap
```

```python
df = pd.read_csv(PROCESSED_DIR / 'cleaned.csv')
df.shape
```

```python
# Black-minus-White gap and ratio for every year
gap = black_white_gap(df)
gap.describe().round(2)
```

```python
# each year gives a Black rate and a White rate, so a paired t-test fits the disparity
t, p = stats.ttest_rel(gap['black'], gap['white'])
print(f'paired t = {t:.2f}, p = {p:.3e}')
print(f'mean gap = {gap["difference"].mean():.2f} per 1,000')
```

```python
# how strongly does the rate move with the year, within each race?
for race, g in df.groupby('race'):
    r, p = stats.pearsonr(g['year'], g[RATE_COL])
    print(f'{race:6} Pearson r = {r:.3f}, p = {p:.2e}')
```

## Notes

- The gap is large, persistent, and highly significant.
- Year and rate are strongly negatively correlated, but the relationship is curved - a linear correlation understates it (see notebook 04).
