# 04 — Modeling the Decline

Regress the rate on the year to characterize the long-run decline, on both a linear and a log scale (the log scale captures the roughly exponential drop).

**Input**: `data/processed/cleaned.csv`

```python
# add the project folder to the path so we can import our code in src/
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
```

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf

from src.config import PROCESSED_DIR, RATE_COL
```

```python
df = pd.read_csv(PROCESSED_DIR / 'cleaned.csv')
# center the year so the intercept is interpretable
df['year_c'] = df['year'] - df['year'].mean()
df['log_rate'] = np.log(df[RATE_COL])
df.head()
```

```python
# simple linear regression of rate on year (per race via an interaction)
linear = smf.ols(f'{RATE_COL} ~ year_c * race', data=df).fit()
print(linear.summary())
```

```python
# log-scale regression: a straight line here means a constant % decline per year
logfit = smf.ols('log_rate ~ year_c * race', data=df).fit()
print(logfit.summary())
print('R-squared linear:', round(linear.rsquared, 3), '| log:', round(logfit.rsquared, 3))
```

## Notes

- The log-scale fit should explain more variance - the decline is closer to exponential than linear.
- Yearly rates are autocorrelated, so the standard errors here are optimistic; this is a descriptive trend model, not a forecast.
