# 04 — Modeling the Decline (CDC)

Regress the rate on the year on linear and log scales (log captures the ~exponential drop).

**Input**: CDC tidy data

```python
# add the project folder to the path so we can import our code in src/
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
```

```python
import numpy as np
import statsmodels.formula.api as smf

from src.data.sources import cdc
```

```python
df = cdc.tidy()
df['year_c'] = df['year'] - df['year'].mean()  # center the year
df['log_imr'] = np.log(df['imr'])
df.head()
```

```python
# linear regression of rate on year, separate slope per race via the interaction
linear = smf.ols('imr ~ year_c * entity', data=df).fit()
print(linear.summary())
```

```python
# log scale: a straight line means a constant % decline per year
logfit = smf.ols('log_imr ~ year_c * entity', data=df).fit()
print(logfit.summary())
print('R^2 linear:', round(linear.rsquared, 3), '| log:', round(logfit.rsquared, 3))
```

## Notes

- The log fit should explain more variance (decline closer to exponential).
- Yearly rates are autocorrelated, so SEs are optimistic — this is descriptive, not a forecast.
